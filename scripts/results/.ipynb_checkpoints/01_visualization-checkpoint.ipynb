# Travelling waves follow a connectivity in-strength gradient

In this notebook, I show that travelling waves emerge and propagate along a connectivity in-strength gradient in a 2D network of 400 Kuramoto oscillators with an intrinsic frequency of 10 Hz. Additionally, I show that this in-strength gradient produces an effective frequency gradient. The network connectivity is based on a distance-dependent connection probability (exponential distribution). The total connection probability was chosen to be around 10%. After oscillators were randomly connected, connection strengths were partitioned randomly to produce a specific in-strength gradient. This procedure ensures that there is no bias in connection strengths along a certain direction. Signals between oscillators have a distance-dependent delay determined by a specific conduction speed (3 m/s, x- and y-extent of network was 100 mm). The initial phases of the Kuramoto oscillators were random uniform (TVB default). The system was evolved for 10 s using the Runge-Kutta 4th order method with an integration step size of 1.0 ms.

To understand propagation of activity, I used a discrete differential operator defined on the 2D mesh to calculate the phase gradients, which should vary systematically if waves emerge. Next, I compute the phase gradient directionality (PGD), a measure devoped by <a href='#Rubino2006'>Rubino et al. (2006)</a> that indicates how well phase gradients are aligned across a 2D mesh at a certain moment in time. If the measure is 1, perfect plane waves propagate across the 2D model and if it is 0, the activity is random.

```python
import os, sys
import yaml
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns

sys.path.append('../../modules/')
from visualization import *

plt.rcParams['image.cmap'] = 'cividis'
```

### Get configuration

```python
# Read Configuration
try:
    configuration_path = "../../configuration/01_configuration.yaml"
    with open(configuration_path, "r") as ymlfile:
        config = yaml.load(ymlfile, Loader=yaml.BaseLoader)
except BaseException as e:
    print("Error: Specify correct path to yaml configuration file.")
    raise


# Simulation Parameters
# ---------------------
data_path = "/Users/dk/Documents/Charite/PhD/travelingwaves/data"

integration_step_size = float(config["integration_step_size"])  # ms
simulation_duration = float(config["simulation_duration"])  # ms
initial_transient = float(config["initial_transient"])  # ms
initial_transient_samples = int(initial_transient / integration_step_size)
number_of_timesteps = int((simulation_duration-initial_transient)/integration_step_size)


# Model Parameters
# ----------------
# network parameters
nx = int(config["nx"])  # number of regions along x-dimension
ny = int(config["ny"])  # number of regions along y-dimension

N = nx*ny  # number of regions

x_extent = float(config["x_extent"])  # mm, extent of x-dimension
y_extent = float(config["y_extent"])  # mm, extent of y-dimension

# create node positions
y = np.linspace(0, y_extent, ny)
x = np.linspace(0, x_extent, nx)
pos = np.zeros((N,2))
pos[:,:] = np.array(np.meshgrid(x, y)).T.reshape(-1,2)

# get specified instrength gradient
instrength_gradient = np.array(config["instrength_gradient"], dtype=float)  # [x,y] - e.g. [1,0] is a vector pointing towards the right
instrength_gradient_offset = float(config["instrength_gradient_offset"])
```

### Show example data

**Instrength gradient example**

```python
# create spatial in-strength gradient
instrength_gradient_field = instrength_gradient @ pos.transpose() + instrength_gradient_offset
```

```python
# plot instrengths on 2D sheet
plt.figure(figsize=(4,3.5))
plt.scatter(pos[:,0], pos[:,1], c = instrength_gradient_field, edgecolors="black", s=50)
plt.box(on=None)
ax = plt.gca()
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_xticks([])
ax.set_yticks([])
cbar = plt.colorbar(shrink=0.97) #, ticks=[2, 12])
cbar.ax.set_ylabel('In-strength (a.u.)')

#plt.savefig(os.path.join(data_path, '01_results/01_instrength-gradient.png'), dpi=300, transparent=False, bbox_inches='tight')
#plt.savefig(os.path.join(data_path, '01_results/01_instrength-gradient.svg'), dpi=300, transparent=False, bbox_inches='tight')
```

**Figure**. This figure shows the in-strengths of each oscillator in the 2D network. The hardcoded in-strength gradient from the left to the right is visible.

#### Synchronization

```python
order_parameter = np.zeros(100)
order_parameter_control = np.zeros(100)
for sid in range(100):
    # load phase
    phase = np.load(os.path.join(data_path, '01_simulations', f'01_simulation_{sid}.npy')).T
    phase_control = np.load(os.path.join(data_path, '02_simulations', f'02_simulation_{sid}.npy')).T
    
    # remove transient
    phase = phase[:, np.arange(initial_transient_samples, number_of_timesteps+initial_transient_samples, integration_step_size).astype(int)]
    phase_control = phase_control[:, np.arange(initial_transient_samples, number_of_timesteps+initial_transient_samples, integration_step_size).astype(int)]

    # compute kuramoto order parameter
    order_parameter[sid] = np.mean(abs(np.mean(np.exp(1j*phase), axis=0)))
    order_parameter_control[sid] = np.mean(abs(np.mean(np.exp(1j*phase_control), axis=0)))
```

```python
# synchronization
dpi = 300
fig, ax = plt.subplots(1, figsize=(2256/dpi/3, 912/dpi/2))

sns.histplot(order_parameter, ax=ax, color='#798897')

plt.ylabel('Number of\n simulations')
plt.xlabel('Synchronization\n order parameter (a.u.)')
sns.despine()

plt.savefig(os.path.join(data_path, '01_results/01_synchronization.png'), dpi=300, transparent=True, bbox_inches='tight')
plt.savefig(os.path.join(data_path, '01_results/01_synchronization.svg'), dpi=300, transparent=True, bbox_inches='tight')
```

```python
# synchronization control
dpi = 300
fig, ax = plt.subplots(1, figsize=(2256/dpi/3, 912/dpi/2))

sns.histplot(order_parameter_control, ax=ax, color='#798897')

plt.ylabel('Number of\n simulations')
plt.xlabel('Synchronization\n order parameter (a.u.)')
sns.despine()

plt.savefig(os.path.join(data_path, '01_results/02_synchronization.png'), dpi=300, transparent=True, bbox_inches='tight')
plt.savefig(os.path.join(data_path, '01_results/02_synchronization.svg'), dpi=300, transparent=True, bbox_inches='tight')
```

**Travelling waves example**

```python
# load data
simulation_id = 10

phase = np.load(os.path.join(data_path, f"01_simulations/01_simulation_{simulation_id}.npy"))
phase = phase[initial_transient_samples:].copy()

# map phase to 2D
phase_2d = np.mod(phase.reshape(number_of_timesteps, nx, ny).transpose((2,1,0)), np.pi*2)

phase_gradient = np.load(os.path.join(data_path, f"01_analysis/phase_gradients_{simulation_id}.npy"))[:,:,initial_transient_samples:]
tw_direction = -phase_gradient
avg_phase_gradient = np.mean(tw_direction, axis=1).T

activity = np.real(np.exp(1j*phase_2d))
```

```python
# wave example plots
number_of_images = 7
time_start = 2080  # ms
time_end = 2200  # ms
samples = (np.linspace(time_start, time_end, number_of_images)/integration_step_size).astype(int)
xlabels = [f"{s - np.min(samples)} ms" for s in samples] 

cmap = sns.color_palette('mako', as_cmap=True)

plot_imshow_quiver_timeseries(activity, avg_phase_gradient, samples, xlabels, 'Gradient', 'Activity (a.u.)',
                              vmin=-1,
                              vmax=1,
                              cbticks=[-1, 0, 1], 
                              cbticklabels=['-1', '0', '1'],
                              cmap=cmap
                             )

plt.savefig(os.path.join(data_path, '01_results/01_example-wave.png'), dpi=300, transparent=False, bbox_inches='tight')
plt.savefig(os.path.join(data_path, '01_results/01_example-wave.svg'), dpi=300, transparent=False, bbox_inches='tight')
```

**Figure**. This figure shows the progression of a travelling wave from left to right. A white arrow indicating the average detected direction across locations at the specific timepoint is superimposed. Clearly, the travelling wave propagates along the specified in-strength gradient.

```python
# generate movie of phase maps and average direction vector
ani = imshow_quiver_movie(activity, avg_phase_gradient, range(0, number_of_timesteps), 10, 'Activity (a.u.)',
                          vmin=-1,
                          vmax=1,
                          cbticks=[-1, 0, 1], 
                          cbticklabels=['-1', '0', '1'],
                          cmap=cmap
                         )

ani.save(os.path.join(data_path,'01_results/01_example.mp4'))
```

**Effective instantaneous frequency example**

Here, effective frequency is the measured frequency of the Kuramoto oscillators as opposed to their specified intrinsic oscillation frequency which was set to exactly 10 Hz for all oscillators.

```python
# compute instantaneous frequency
instantaneous_frequency = np.diff(phase, axis=0) / (2.0*np.pi*integration_step_size) * 1e3  # scaling because TVB simulates in milliseconds
instantaneous_frequency_2d = instantaneous_frequency.reshape(number_of_timesteps-1, nx, ny).transpose((2,1,0))
```

```python
# effective frequency example plots

cmap = sns.color_palette('crest_r', as_cmap=True)

plot_imshow_timeseries(instantaneous_frequency_2d, samples, xlabels, 'Gradient', 'Effective frequency (Hz)',cmap=cmap)

plt.savefig(os.path.join(data_path, '01_results/01_example-effective-frequency.png'), dpi=300, transparent=False, bbox_inches='tight')
plt.savefig(os.path.join(data_path, '01_results/01_example-effective-frequency.svg'), dpi=300, transparent=False, bbox_inches='tight')
```

**Figure**. Plotting the instantaneous effective frequency for the same travelling wave example shows that the in-strength gradient produces an effective frequency gradient that appears to remain constant over time. The effective frequency is mostly lower than the specified intrinsic frequency of 10 Hz.

**Proportion of timepoints exhibiting travelling waves**

Here, I compute the proportion of timepoints that exhibit travelling waves across all 100 trials for the gradient and control model. A travelling wave was defined by crossing a phase gradient directionality (PGD) threshold of 0.5 accroding to <a href='#Rubino2006'>Rubino et al. (2006)</a>.

```python
# load data
n_trials = 100
PGD = []
PGD_ctrl = []

for i in range(n_trials):
    PGD.append(np.load(os.path.join(data_path, f"01_analysis/pgd_{i}.npy"))[initial_transient_samples:])
    PGD_ctrl.append(np.load(os.path.join(data_path, f"02_analysis/pgd_{i}.npy"))[initial_transient_samples:])
```

```python
# compute proportion of timepoints across trials that exhibit travelling waves
detection_threshold = 0.5

proportion_travelling_waves = [np.sum(PGD[i]>detection_threshold)/number_of_timesteps for i in range(len(PGD))]
proportion_travelling_waves_ctrl = [np.sum(PGD_ctrl[i]>detection_threshold)/number_of_timesteps for i in range(len(PGD_ctrl))]
```

```python
print("Mean proportion of traveling waves: %f +- %f" % (np.mean(proportion_travelling_waves), np.std(proportion_travelling_waves)))
print("Mean proportion of traveling waves - Control: %f +- %f" % (np.mean(proportion_travelling_waves_ctrl), np.std(proportion_travelling_waves_ctrl)))
```

```python
fig, ax = plt.subplots(1, figsize=(1.8,2.5), dpi=300)
c = plt.cm.cividis(0.95)
ax.boxplot(proportion_travelling_waves, positions=[0], notch=True, patch_artist=True,
           boxprops=dict(facecolor=c, color=c),
           capprops=dict(color=c),
           whiskerprops=dict(color=c),
           flierprops=dict(color=c, markeredgecolor=c),
           medianprops=dict(color=c)
          )
c = plt.cm.cividis(0.05)
ax.boxplot(proportion_travelling_waves_ctrl, positions=[0.4], notch=True, patch_artist=True,
           boxprops=dict(facecolor=c, color=c),
           capprops=dict(color=c),
           whiskerprops=dict(color=c),
           flierprops=dict(color=c, markeredgecolor=c),
           medianprops=dict(color=c)
          )
ax.set_xticklabels(['Gradient', 'Uniform'], rotation=45)
ax.set_ylabel("Proportion of timepoints\n exhibiting travelling waves")
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.xlim([-0.2, 0.6])

plt.savefig(os.path.join(data_path, '01_results/proportion-travelling-waves-boxplot.png'), dpi=300, transparent=False, bbox_inches='tight')
plt.savefig(os.path.join(data_path, '01_results/proportion-travelling-waves-boxplot.svg'), dpi=300, transparent=False, bbox_inches='tight')
```

**Figure**. The boxplots show that travelling waves almost never emerge in the uniform model, while the gradient model expresses travelling waves 60% of the time across all trials.

**Polar histogram of average phase-direction**

```python
# load data
n_trials = 100
direction = []
direction_ctrl = []

for i in range(n_trials):
    direction.append(-np.load(os.path.join(data_path, f"01_analysis/phase_gradients_{i}.npy"))[:-1,:,initial_transient_samples:])
    direction_ctrl.append(-np.load(os.path.join(data_path, f"02_analysis/phase_gradients_{i}.npy"))[:-1,:,initial_transient_samples:])
```

```python
# compute the average direction across loactions and timepoints
avg_direction = np.array([np.mean(direction[i], axis=(1,2)) for i in range(n_trials)])
avg_direction_ctrl = np.array([np.mean(direction_ctrl[i], axis=(1,2)) for i in range(n_trials)])

# compute polar angle for each trial
angles = np.angle(avg_direction[:,0]+1j*avg_direction[:,1])
angles_ctrl = np.angle(avg_direction_ctrl[:,0]+1j*avg_direction_ctrl[:,1])
```

```python
# polar histogram of phase-directions across trials
polar_histogram(angles, angles_ctrl, bin_width=2*np.pi/36, ylabel="Wavedirection", xticklabels=['x', 'y'], legend=['Gradient', 'Uniform'])

plt.savefig(os.path.join(data_path, '01_results/direction-polar-histogram.png'), dpi=300, transparent=False, bbox_inches='tight')
plt.savefig(os.path.join(data_path, '01_results/direction-polar-histogram.svg'), dpi=300, transparent=False, bbox_inches='tight')
```

**Figure**. This figure shows a polar histogram of the average phase-directions (across timepoints and locations) across 100 trials for the gradient and control condition. While the control condition with uniform in-strength shows a uniform distribution of directions, the model with a hardcoded in-strength gradient shows a clear bias along the x-dimension that follows the in-strength gradient. Notably, this polar histogram only shows the average phase-direction but does not indicate the presence of travelling waves. Detecting travelling waves requires alignment of phase-gradients across the 2D network, such an alignment could not be detected with the phase gradient directionality (PGD, <a href='#Rubino2006'>Rubino et al., 2006</a>) in the control simulation. However, PGD was pronounced in the simulations with an in-strength gradient. Therefore, while models with an in-strength gradient exhibit strong travelling waves following that very same gradient, control simulations with a uniform in-strength do not produce travelling waves but synchronize over time.

```python
# Test if wave directions are non-uniform
from pingouin import circ_rayleigh

print("Gradient: \n\tz-statistic: %.3f, p: %.3f" % circ_rayleigh(angles))
print("Control: \n\tz-statistic: %.3f, p: %.3f" % circ_rayleigh(angles_ctrl))
```

**Statistics**: While the model with an instrength gradient is non-uniform the control model exhibits traveling waves in all directions.
