# Travelling waves follow a connectivity in-strength gradient in a full brain network model

In this notebook, I show that travelling waves emerge and propagate along a connectivity in-strength gradient in a full brain network model consisting of 1000 Kuramoto oscillators with an intrinsic frequency of 10 Hz. Additionally, I show that this in-strength gradient produces an effective frequency gradient. The network connectivity is based on the structural connectivity between the regions of the Schaefer atlas (<a href='#Schaefer2018'>Schaefer et al., 2018</a>) estimated by tractography from 776 subjects of the Human Connectome Project S900 release (for details see [20_visualization](./20_visualization.ipynb)). Signals between oscillators propagate with a tract-length-dependent delay determined by a 3 m/s conduction speed. The initial phases of the Kuramoto oscillators were random uniform (TVB default). The system was evolved for 10 s using the Runge-Kutta 4th order method with an integration step size of 1.0 ms.

To understand propagation of activity, I first computed spatial differential operators using the meshfree generalized finite difference method (mGFDM) presented in <a href='#Suchde2019'>Suchde and Kuhnert (2019)</a>. Their method allows the calculation of a spatial differential operator on 2D surfaces embedded in 3-space (e.g. cortical surface in 3D). Here, I compute the differential operator on a 2D manifold spanned by the locations of the regions defined by the Schaefer atlas to calculate the phase, frequency, and instrength gradients for each point in space and time.

Next, I computed the angular similiarity (a measure of how well vectors are aligned) between the negative-phase and instrength gradients to show that travelling wave propagation direction is aligned with the structural connectivity instrength gradient. Furthermore, I show that this relationship holds for the effective frequency and the negative phase gradients.

**Travelling waves example**

```python
# load data
simulation_id = 1
phase = np.load(os.path.join(data_path, f'30_simulations/30_simulation_{simulation_id}.npy'))

# wrap phase
phase_wrapped = np.mod(phase, np.pi*2)
```

```python
samples_interval = 20
samples = np.arange(4680,4800,samples_interval)
n_samples = len(samples)
fig, ax = plt.subplots(n_samples, 3, subplot_kw={'projection':'3d'}, figsize=(page_width/2/dpi,800*n_samples/2/dpi), dpi=dpi)

elev = [0, 90, 0]
azim = [180, 270, 0]

zdirs = (None, 'x', 'y', 'z', (1, 1, 0), (1, 1, 1))
    
for j in range(n_samples):
    tp = samples[j]

    for i in range(3):
        im = ax[j,i].scatter(pos[:,0], pos[:,1], pos[:,2], s=20, c=phase_wrapped[tp,:]-np.pi,
                             cmap='twilight',
                             vmin=-np.pi,
                             vmax=np.pi,
                             alpha=1.0
                            )
        ax[j,i].view_init(elev[i], azim[i])
        
        ax[j,i].set_axis_off()
        ax[j,i].set_box_aspect((np.ptp(pos[:,0]), np.ptp(pos[:,1]), np.ptp(pos[:,2])), zoom=1.5)

    ax[j,0].set_title(f'{j*samples_interval*integration_step_size} ms', loc='left', pad=0.1)
        
fig.subplots_adjust(bottom=0.05, wspace=0.)
cbar_ax = fig.add_axes([0.31, 0., 0.41, 0.05/n_samples])
cbar = fig.colorbar(im, cax=cbar_ax, orientation='horizontal')
cbar.ax.tick_params(size=0)
cbar.set_ticks([-np.pi, np.pi])
cbar.set_ticklabels(['$-\pi$', '$\pi$'])
cbar.set_label('Phase (rad)')
cbar.outline.set_visible(False)


#plt.savefig(os.path.join(figure_path, f'30_phase_timeseries.png'), dpi=300, transparent=False, bbox_inches='tight')
```

**Figure**. This timeseries shows a large-scale travelling wave propagating from posterior to anterior regions.

**Effective instantaneous frequency example**

Here, effective frequency is the measured frequency of the Kuramoto oscillators as opposed to their specified intrinsic oscillation frequency which was set to exactly 10 Hz for all oscillators.

```python
# compute instantaneous frequency
instantaneous_frequency = np.diff(phase, axis=0) / (2.0*np.pi*integration_step_size) * 1e3  # scaling because TVB simulates in milliseconds
```

```python
np.min(instantaneous_frequency)
```

```python
samples_interval = 20
samples = np.arange(3800,3920,samples_interval)
n_samples = len(samples)
fig, ax = plt.subplots(n_samples, 3, subplot_kw={'projection':'3d'}, figsize=(page_width/2/dpi,800*n_samples/2/dpi), dpi=dpi)

elev = [0, 90, 0]
azim = [180, 270, 0]

zdirs = (None, 'x', 'y', 'z', (1, 1, 0), (1, 1, 1))
    
for j in range(n_samples):
    tp = samples[j]

    for i in range(3):
        im = ax[j,i].scatter(pos[:,0], pos[:,1], pos[:,2], s=20, c=instantaneous_frequency[tp,:]-np.pi, alpha=1.0)
        ax[j,i].view_init(elev[i], azim[i])
        
        ax[j,i].set_axis_off()
        ax[j,i].set_box_aspect((np.ptp(pos[:,0]), np.ptp(pos[:,1]), np.ptp(pos[:,2])), zoom=1.5)

    ax[j,0].set_title(f'{j*samples_interval*integration_step_size} ms', loc='left', pad=0.1)
        
fig.subplots_adjust(bottom=0.05, wspace=0.)
cbar_ax = fig.add_axes([0.31, 0., 0.41, 0.05/n_samples])
cbar = fig.colorbar(im, cax=cbar_ax, orientation='horizontal')
#cbar.ax.tick_params(size=0)
#cbar.set_ticks([-np.pi, np.pi])
#cbar.set_ticklabels(['$-\pi$', '$\pi$'])
cbar.set_label('Effective frequency (Hz)')
cbar.outline.set_visible(False)


#plt.savefig(os.path.join(figure_path, f'30_effective-frequency_timeseries.png'), dpi=300, transparent=False, bbox_inches='tight')
```

**Figure**. Plotting the instantaneous effective frequency for the same travelling wave example shows that the in-strength gradient produces an effective frequency gradient that appears to remain constant over time, however, the frequency range appears to shift. The effective frequency is mostly lower than the specified intrinsic frequency of 10 Hz.

**Example angular similarity**

```python
# load data
angular_similarity = np.load(os.path.join(data_path, '30_analysis', f'angular_similarity_instrength_phase_{simulation_id}.npy'))
```

```python
# angular similarity distribution across regions for each timepoint
fig, ax = plt.subplots(1,1, figsize=(page_width/3/dpi,500/dpi), sharey=True, dpi=dpi)

timehist = np.array([np.histogram(angular_similarity[tp], bins=np.arange(-1, 1.1, 0.1))[0]*100/number_of_regions for tp in range(number_of_timesteps)])

im = ax.imshow(timehist.transpose(), aspect='auto', extent=[0, number_of_timesteps/integration_step_size, -1, 1], origin='lower')
ax.set_xlabel('Time (ms)')
ax.set_ylabel('Angular similarity (a.u.)')
ax.set_yticks([-1, 0, 1])
ax.set_frame_on(False)

cbar = plt.colorbar(im, label='Percent of regions')
cbar.outline.set_visible(False)

#plt.savefig(os.path.join(figure_path, f'30_angular-similarity_instrength-phase_heatmap.png'), dpi=300, transparent=False, bbox_inches='tight')
```

**Figure**. Time evolution of angular similarity between instrength and negative phase gradients. This heatmap shows that the instrength and negative phase gradients of many brain regions strongly align after a brief initial transient. About 20% of all regions exhibit very strong angular similarity during most of the simulation.

```python
# central tendency of angular similarity across regions over time
median_angsim = np.median(angular_similarity, axis=1)
mean_angsim = np.mean(angular_similarity, axis=1)
t = np.arange(0, number_of_timesteps*integration_step_size, integration_step_size)

# plot results
fig, ax = plt.subplots(1,1, figsize=(page_width/3/dpi,500/dpi), sharey=True, dpi=dpi)

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.plot(t, median_angsim, c='black')
ax.plot(t, mean_angsim, c='gray')
ax.legend(['median', 'mean'], frameon=False)
ax.set_xlabel("Time (ms)")
ax.set_ylabel("Angular similarity (a.u.)")

#plt.savefig(os.path.join(figure_path, f'30_angular-similarity_instrength-phase_central-tendency.png'), dpi=300, transparent=False, bbox_inches='tight')
```

**Figure**. Time evolution of central tendency of angular similarity across regions. The mean and median angular similarity across regions reaches positive intermediate values after an initial transient. This indicates that the instrength and negative phase gradients are strongly aligned in most regions during the majority of the simulation.

**Average effective frequency**

```python
# load simulations
phases = []

for i in simulation_ids:
    phases.append(np.load(os.path.join(data_path, '30_simulations', f'30_simulation_{i}.npy')))
```

```python
# compute effective instantaneous frequency
instantaneous_frequency = []

for i in range(len(simulation_ids)):
    instantaneous_frequency.append(np.diff(phases[i], axis=0) / (2.0*np.pi*integration_step_size) * 1e3)  # scaling because TVB simulates in milliseconds
```

```python
# mean across simulations
mean_instantaneous_frequency = np.mean(instantaneous_frequency, axis=0)
```

```python
# average across time
start_sample = 1000  # initial transient to discard for temporal averaging of frequency
end_sample = 9999
average_effective_frequency = np.median(mean_instantaneous_frequency[start_sample:end_sample,:], axis=0)

#q = [2.5, 97.5]
q = [0, 100]
p = np.percentile(average_effective_frequency, q)
average_effective_frequency[(average_effective_frequency<p[0])|(average_effective_frequency>p[1])] = np.nan
```

```python
# plot average effective frequency
fig, ax = plt.subplots(1, 3, subplot_kw={'projection':'3d'}, figsize=(page_width/dpi,500/dpi), dpi=dpi)

elev = [0, 90, 0]
azim = [180, 270, 0]

for i in range(3):
    im = ax[i].scatter(pos[:,0], pos[:,1], pos[:,2], s=20, 
                       c=average_effective_frequency, 
                       alpha=1.0, 
                       plotnonfinite=True
                      )
    ax[i].view_init(elev[i], azim[i])
    
    ax[i].set_axis_off()
    ax[i].set_box_aspect((np.ptp(pos[:,0]), np.ptp(pos[:,1]), np.ptp(pos[:,2])), zoom=1.5)
    
fig.subplots_adjust(bottom=0.2, wspace=-0.6)
cbar_ax = fig.add_axes([0.40, 0.1, 0.23, 0.05])
cbar = fig.colorbar(im, cax=cbar_ax, orientation='horizontal')
cbar.set_label('Average effective frequency (Hz)')
cbar.outline.set_visible(False)

#plt.savefig(os.path.join(figure_path, '30_average-effective-frequency.png'), dpi=300, transparent=False, bbox_inches='tight')
```

**Figure** The effective frequency averaged across simulations and timepoints reveals a clear spatial pattern. Effective frequency is lowest in the frontal lobe, central and occipital regions oscillate at intermediate frequencies while the temporal lobes exhibit the highest frequency oscillations. This pattern is so regular that I should check if this can be an artifact of some sort.
