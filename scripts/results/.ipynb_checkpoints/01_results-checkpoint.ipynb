# Travelling waves follow a connectivity in-strength gradient in a 2D network model

```python
import os, sys
import yaml
import numpy as np
import pandas as pd
import matplotlib.pylab as plt

from scipy.stats import spearmanr, multivariate_normal
from scipy import spatial
from palettable.cmocean.sequential import Ice_20
from palettable.scientific.sequential import Davos_20, LaPaz_20, Turku_20

import cmasher as cmr

from modules.helpers import *
from modules.wave_detection_methods import *
from modules.visualization import *

cmap_as = cmr.eclipse
cmap_waves = Ice_20.mpl_colormap
cmap_strength = Davos_20.mpl_colormap
cmap_fq = LaPaz_20.mpl_colormap

c1 = '#3baa70'
c2 = '#703baa'

dpi = 300
plt.rcParams.update({'font.size': 8})
```

```python
# Read Configuration
try:
    configuration_path = "/Users/dk/Documents/Charite/PhD/travelingwaves/configuration/01_configuration.yaml"
    with open(configuration_path, "r") as ymlfile:
        config = yaml.load(ymlfile, Loader=yaml.BaseLoader)
except BaseException as e:
    print("Error: Specify correct path to yaml configuration file.")
    raise

try:
    configuration_path = "/Users/dk/Documents/Charite/PhD/travelingwaves/configuration/01_analysis_configuration.yaml"
    with open(configuration_path, "r") as ymlfile:
        config_analysis = yaml.load(ymlfile, Loader=yaml.BaseLoader)
except BaseException as e:
    print("Error: Specify correct path to yaml configuration file.")
    raise

    
# Simulation Parameters
# ---------------------
data_path = "/Users/dk/Documents/Charite/PhD/travelingwaves/data"
experiment_id = config["experiment_id"]

number_of_simulations = 100

integration_step_size = float(config["integration_step_size"])  # ms
simulation_duration = float(config["simulation_duration"])  # ms
initial_transient = float(config["initial_transient"])  # ms
initial_transient_samples = int(initial_transient/integration_step_size)  # samples
number_of_timesteps = int((simulation_duration-initial_transient)/integration_step_size)

downsampling_factor = float(config_analysis["downsampling_factor"])  # a.u.
integration_step_size_downsampled = integration_step_size * downsampling_factor
number_of_timesteps_downsampled = int((simulation_duration-initial_transient)/integration_step_size_downsampled)


# Model Parameters
# ----------------
# network parameters
nx = int(config["nx"])  # number of regions along x-dimension
ny = int(config["ny"])  # number of regions along y-dimension

N = nx*ny  # number of regions

x_extent = float(config["x_extent"])  # mm, extent of x-dimension
y_extent = float(config["y_extent"])  # mm, extent of y-dimension

# create node positions
y = np.linspace(0, y_extent, ny)
x = np.linspace(0, x_extent, nx)
pos = np.zeros((N,2))
pos[:,:] = np.array(np.meshgrid(x, y)).T.reshape(-1,2)

# prepare mesh
v = np.c_[pos, np.zeros(len(pos))]
f = spatial.Delaunay(pos).simplices


# Analysis Parameters
# -------------------
significance_level = float(config_analysis["significance_level"])
```

### Visualize instrength distribution

```python
# create spatial in-strength gradient
sink_pos = np.array(config["sink_pos"])
source_pos = np.array(config["source_pos"])
singularity_width = float(config["singularity_width"])

sink = multivariate_normal(sink_pos, cov=singularity_width).pdf(pos)  # generate sink gaussian
source = multivariate_normal(source_pos, cov=singularity_width).pdf(pos)  # generate source gaussian

gradient = (sink - source)  # combine gaussians to create gradient
instrength = (gradient - gradient.min()) * 2 / (gradient.max() - gradient.min()) - 1  # min-max normalized gradient
```

```python
from mpl_toolkits.axes_grid1.anchored_artists import AnchoredSizeBar
import matplotlib.font_manager as fm

fontprops = fm.FontProperties(size=7)

fig, ax = plt.subplots(1, figsize=(1.2, 1.2), dpi=300)

scalebar = AnchoredSizeBar(ax.transData,
                           20, '20 mm', 
                           'lower right', 
                           color='black',
                           borderpad=0.25,
                           frameon=False,
                           size_vertical=1,
                           fontproperties=fontprops)

ax.add_artist(scalebar)

im = ax.imshow(instrength.reshape(nx,ny).T, origin='lower', cmap=cmap_strength, aspect=1, extent=[0, 140, 0, 140])
ax.set_title('Gradient')
ax.set_yticklabels([])
ax.set_xticklabels([])
ax.set_yticks([])
ax.set_xticks([])
ax.set_xlabel('X position')
ax.set_ylabel('Y position')

#cbar_ax = fig.add_axes([0.125, 0., 0.775, 0.05])
cbar_ax = fig.add_axes([.95, 0.11, 0.05, 0.77])

cbar = fig.colorbar(im, ax=ax, cax=cbar_ax, pad=0.02, orientation='vertical')
cbar.set_label('Instrength (a.u.)')
cbar.set_ticks(ticks=[instrength.min(), instrength.max()])
cbar.set_ticklabels(['low', 'high'], va='center')
cbar.ax.tick_params(size=0, rotation=90)

plt.savefig(os.path.join(data_path, '01_results/01_average_instrength.png'), dpi=300, transparent=False, bbox_inches='tight')
plt.savefig(os.path.join(data_path, '01_results/01_average_instrength.svg'), dpi=300, transparent=False, bbox_inches='tight')
```

### Wave analysis

```python
proportion_waves = np.zeros(number_of_simulations)
proportion_div_waves = np.zeros(number_of_simulations)
proportion_rot_waves = np.zeros(number_of_simulations)

wave_potential_div_sum = np.zeros(N)

proportion_corr = np.zeros(number_of_simulations)

corr = np.zeros((number_of_simulations, number_of_timesteps_downsampled))
corr_sim = np.zeros(number_of_simulations)

instantaneous_frequency_sum = np.zeros(N)

for sid in range(number_of_simulations):
    # Find significant singularities
    p_div = np.load(os.path.join(data_path, f'01_analysis', f'01_p_div_{sid}.npy'))
    p_rot = np.load(os.path.join(data_path, f'01_analysis', f'01_p_rot_{sid}.npy'))
    
    significant_div = p_div <= significance_level
    significant_rot = p_rot <= significance_level

    # Create wave mask
    wave_mask_div = np.nansum(significant_div, axis=0, dtype=bool)    
    wave_mask_rot = np.nansum(significant_rot, axis=0, dtype=bool)

    proportion_waves[sid] = np.nansum(wave_mask_div | wave_mask_rot) / number_of_timesteps_downsampled


    # Laod potential
    potential_div = np.load(os.path.join(data_path, f'01_analysis_potentials/01_potential_div_{sid}.npy'))
    
    # compute proportion waves
    proportion_div_waves[sid] = np.nansum(wave_mask_div) / number_of_timesteps_downsampled
    
    
    # Compute effective frequency
    # ---------------------------
    # load phase
    phase = np.load(os.path.join(data_path, f'01_simulations', f'01_simulation_{sid}.npy')).T
    phase = phase[:, np.arange(initial_transient_samples, number_of_timesteps+initial_transient_samples, integration_step_size).astype(int)]  # remove transient
    ce_phase = np.exp(1j*phase)  # complex exponential phase
    
    instantaneous_frequency = compute_instantaneous_frequency(ce_phase, integration_step_size*1e-3).T
    instantaneous_frequency = instantaneous_frequency[np.arange(0, number_of_timesteps, integration_step_size_downsampled).astype(int)]  # downsampling

    
    # Stats
    # -----
    if wave_mask_rot.sum():
        proportion_rot_waves[sid] = np.nansum(wave_mask_rot) / number_of_timesteps_downsampled
        
    if wave_mask_div.sum():
        # Sum up time-averaged potential across waves
        wave_potential_div_sum_tmp = np.nanmean(potential_div[:,wave_mask_div], axis=1)
        wave_potential_div_sum = np.nansum([wave_potential_div_sum, wave_potential_div_sum_tmp], axis=0)
    
        # Load instrength-potential correlation
        corr[sid] = np.load(os.path.join(data_path, '01_analysis_potentials', f'01_corr_div_{sid}.npy'))

        # Assess significance of correlation
        p_corr = np.load(os.path.join(data_path, '01_analysis_potentials', f'01_p_corr_{sid}.npy'))

        significant_corr = p_corr <= significance_level  # one-sided hypothesis: instrength is negatively correlated to potentials

        # Proportion of significant correlation during waves
        proportion_corr[sid] = np.nansum(significant_corr & wave_mask_div) / np.sum(wave_mask_div)

        # Sum up correlation means
        corr_sim[sid] = np.nanmean(corr[sid][wave_mask_div & significant_corr])
        
        # instantaneous frequency median
        instantaneous_frequency_sum += np.nanmedian(instantaneous_frequency[wave_mask_div], axis=0)
            
        
# Compute means - left hemi
wave_potential_div_mean = wave_potential_div_sum / number_of_simulations

proportion_waves_median = np.nanmedian(proportion_waves)
proportion_div_waves_median = np.nanmedian(proportion_div_waves)
proportion_rot_waves_median = np.nanmedian(proportion_rot_waves)
proportion_corr_median = np.nanmedian(proportion_corr)

proportion_waves_quartiles = np.nanpercentile(proportion_waves, [25, 75])
proportion_div_waves_quartiles = np.nanpercentile(proportion_div_waves, [25, 75])
proportion_rot_waves_quartiles = np.nanpercentile(proportion_rot_waves, [25, 75])
proportion_corr_quartiles = np.nanpercentile(proportion_corr, [25, 75])

corr_time_avg_mean = np.nanmean(corr_sim)
corr_time_avg_sd = np.nanstd(corr_sim)

corr_mean = np.nanmean(corr, axis=0)
corr_sd = np.nanstd(corr, axis=0)

effective_frequency = instantaneous_frequency_sum / number_of_simulations


# Print stats
print(f"\tMedia (IQR) proportion of wave episodes across simulations: {proportion_waves_median:.2f}  ({proportion_waves_quartiles[0]:.2f} - {proportion_waves_quartiles[1]:.2f})\n"
      f"\tMedian (IQR) proportion of diverging wave episodes across simulation duration: {proportion_div_waves_median:.2f} ({proportion_div_waves_quartiles[0]:.2f} - {proportion_div_waves_quartiles[1]:.2f})\n"
      f"\tMedian (IQR) proportion of rotating wave episodes across simulation duration: {proportion_rot_waves_median:.2f} ({proportion_rot_waves_quartiles[0]:.2f} - {proportion_rot_waves_quartiles[1]:.2f})\n"
      f"\tMedia (IQR) proportion of instrength-guided waves across wave episodes: {proportion_corr_median:.2f}  ({proportion_corr_quartiles[0]:.2f} - {proportion_corr_quartiles[1]:.2f})\n"
      f"\tAverage instrength-potential correlation: {corr_time_avg_mean:.2f} \u00B1 {corr_time_avg_sd:.2f}\n\n"
     )
```

### Proportion traveling waves

```python
# Build proportion TWs dataframe for all experiments
experiment_ids = ['01', '02']
experiments = ['Gradient', 'Uniform']

data = []
for exi, ex_id in enumerate(experiment_ids):
    proportion_waves, proportion_corr = aggregate_proportions_2d(data_path, ex_id)
    
    data.append(np.column_stack([np.repeat(experiments[exi], 100),
                                 proportion_waves*100,
                                 proportion_corr*100
                                ]))

data = np.vstack(data)

df = pd.DataFrame(data, columns=['Experiment', 'Proportion TWs (%)', 'Proportion guided TWs (%)'])
df['Proportion TWs (%)'] = df['Proportion TWs (%)'].astype(float)
df['Proportion guided TWs (%)'] = df['Proportion guided TWs (%)'].astype(float)
```

```python
# Plot Proportion Traveling Waves
sns.set_theme(style="ticks")

plt.figure(figsize=(2,2.5))
plt.subplot(2,1,1)
p = sns.violinplot(df, x='Experiment', y='Proportion TWs (%)', palette=[c1, c2], cut=0, linewidth=1, inner=None)
p.set_xticklabels([])
p.set_xlabel('')
p.set_ylabel('')
p.set_title('Proportion of traveling waves (%)')
plt.grid()

medians_proportion_waves = df.groupby(['Experiment'], sort=False)['Proportion TWs (%)'].median()
voffset_proportion_waves = [-30, 20]
hoffset_proportion_waves = [-0.25, -0.25]

for xtick in p.get_xticks():
    p.text(xtick + hoffset_proportion_waves[xtick], 
           medians_proportion_waves[xtick] + voffset_proportion_waves[xtick], 
           f"{medians_proportion_waves[xtick]:.1f}", 
           horizontalalignment='center',size='x-small',color='black')

plt.subplot(2,1,2)
p = sns.violinplot(df, x='Experiment', y='Proportion guided TWs (%)', palette=[c1, c2], cut=0, linewidth=1, inner=None)
p.set_ylabel('')
p.set_title('Proportion of guided waves (%)')
p.set_xticklabels(p.get_xticklabels(), rotation=35)
p.set_xlabel('')
p.xaxis.set_visible('off')
plt.grid()

medians_proportion_guided_waves = df.groupby(['Experiment'], sort=False)['Proportion guided TWs (%)'].median()
voffset_proportion_guided_waves = [-30, 20]
hoffset_proportion_guided_waves = [-0.25, -0.25]

for xtick in p.get_xticks():
    p.text(xtick + hoffset_proportion_guided_waves[xtick], 
           medians_proportion_guided_waves[xtick] + voffset_proportion_guided_waves[xtick], 
           f"{medians_proportion_guided_waves[xtick]:.1f}", 
           horizontalalignment='center',size='x-small',color='black')

plt.subplots_adjust(hspace=0.5)
sns.despine()

plt.savefig(os.path.join(data_path, f'01_results/proportion_waves.png'), dpi=300, transparent=True, bbox_inches='tight')
plt.savefig(os.path.join(data_path, f'01_results/proportion_waves.svg'), dpi=300, transparent=True, bbox_inches='tight')
```
