# Supplementary - Travelling waves follow a connectivity in-strength gradient

In this notebook, I show supplementary results for the 2D network model.

```python
import os, sys
import yaml
import numpy as np
import matplotlib.pylab as plt

sys.path.append('../../modules/')
from visualization import *

plt.rcParams['image.cmap'] = 'cividis'
```

# In-strength gradient with different direction

Here, I show another example of the 2D network of oscillators with a hardcoded in-strength gradient that is different to the one presented in the main simulations. The in-strength gradient in this example goes from bottom-right to top-left.

```python
# Read Configuration
try:
    configuration_path = "../../configuration/11_configuration.yaml"
    with open(configuration_path, "r") as ymlfile:
        config = yaml.load(ymlfile, Loader=yaml.BaseLoader)
except BaseException as e:
    print("Error: Specify correct path to yaml configuration file.")
    raise


# Simulation Parameters
# ---------------------
data_path = "/Users/dk/Documents/Charite/PhD/waves/waves/data"

integration_step_size = float(config["integration_step_size"])  # ms
simulation_duration = float(config["simulation_duration"])  # ms
number_of_timesteps = int(simulation_duration/integration_step_size)


# Model Parameters
# ----------------
# network parameters
nx = int(config["nx"])  # number of regions along x-dimension
ny = int(config["ny"])  # number of regions along y-dimension

N = nx*ny  # number of regions

x_extent = float(config["x_extent"])  # mm, extent of x-dimension
y_extent = float(config["y_extent"])  # mm, extent of y-dimension

# create node positions
y = np.linspace(0, y_extent, ny)
x = np.linspace(0, x_extent, nx)
pos = np.zeros((N,2))
pos[:,:] = np.array(np.meshgrid(x, y)).T.reshape(-1,2)

# get specified instrength gradient
instrength_gradient = np.array(config["instrength_gradient"], dtype=float)  # [x,y] - e.g. [1,0] is a vector pointing towards the right
instrength_gradient_offset = float(config["instrength_gradient_offset"])
```

### Show example data

**Travelling waves example**

```python
# load data
simulation_id = 10

phase = np.load(os.path.join(data_path, f"checks/01_simulations_different-direction/01_simulation_{simulation_id}.npy"))

# map phase to 2D
phase_2d = np.mod(phase.reshape(number_of_timesteps, nx, ny).transpose((2,1,0)), np.pi*2)

phase_gradient = np.load(os.path.join(data_path, f"checks/01_analysis_different-direction/phase_gradients_{simulation_id}.npy"))
tw_direction = -phase_gradient
avg_phase_gradient = np.mean(tw_direction, axis=1)
```

```python
# wave example plots
number_of_images = 7
time_start = 2040  # ms
time_end = 2160  # ms
samples = (np.linspace(time_start, time_end, number_of_images)/integration_step_size).astype(int)
xlabels = [f"{s - np.min(samples)} ms" for s in samples] 

plot_imshow_quiver_timeseries(phase_2d-np.pi, avg_phase_gradient, samples, xlabels, 'Gradient', 'Phase (rad)',
                              vmin=-np.pi,
                              vmax=np.pi,
                              cbticks=[-np.pi, 0, np.pi], 
                              cbticklabels=['$-\pi$', '0', '$\pi$'])

plt.savefig(os.path.join(data_path, '01_results/checks/01_example-wave_different-direction.png'), dpi=300, transparent=False, bbox_inches='tight')
plt.savefig(os.path.join(data_path, '01_results/checks/01_example-wave_different-direction.svg'), dpi=300, transparent=False, bbox_inches='tight')
```

**Figure**. This figure shows the progression of a travelling wave from bottom-right to top-left in a 2D model with an in-strength gradient from bottom-right to top-left. A white arrow indicating the average detected direction across locations at the specific timepoint is superimposed.

```python
# generate movie of phase maps and gradient vectors
ani = imshow_quiver_movie(phase_2d-np.pi, avg_phase_gradient, range(0, number_of_timesteps), 10, 'Phase (rad)',
                          vmin=-np.pi,
                          vmax=np.pi,
                          cbticks=[-np.pi, 0, np.pi], 
                          cbticklabels=['$-\pi$', '0', '$\pi$']
                         )

ani.save(os.path.join(data_path,'01_results/checks/01_example_different-direction.mp4'))
```

```python
# compute instantaneous frequency
instantaneous_frequency = np.diff(phase, axis=0) / (2.0*np.pi*integration_step_size) * 1e3  # scaling because TVB simulates in milliseconds
instantaneous_frequency_2d = instantaneous_frequency.reshape(number_of_timesteps-1, nx, ny).transpose((2,1,0))
```

```python
# effective frequency example plots
plot_imshow_timeseries(instantaneous_frequency_2d, samples, xlabels, 'Gradient', 'Effective frequency (Hz)',cmap="cividis")

plt.savefig(os.path.join(data_path, '01_results/checks/01_example-effective-frequency_different-direction.png'), dpi=300, transparent=False, bbox_inches='tight')
plt.savefig(os.path.join(data_path, '01_results/checks/01_example-effective-frequency_different-direction.svg'), dpi=300, transparent=False, bbox_inches='tight')
```

**Figure**. Plotting the instantaneous effective frequency for the same travelling wave example shows that the in-strength gradient produces an effective frequency gradient that appears to remain constant over time. The effective frequence is mostly lower than the specified intrinsic frequenc of 10 Hz.

**Proportion of timepoints exhibiting travelling waves**

Here, I compute the proportion of timepoints that exhibit travelling waves across all 100 trials for the gradient and control model. A travelling wave was defined by crossing a phase gradient directionality (PGD) threshold of 0.5 accroding to <a href='#Rubino2006'>Rubino et al. (2006)</a>.

```python
# load data
n_trials = 100
PGD = []

for i in range(n_trials):
    PGD.append(np.load(os.path.join(data_path, f"checks/01_analysis_different-direction/pgd_{i}.npy")))
```

```python
# compute proportion of timepoints across trials that exhibit travelling waves
detection_threshold = 0.5

proportion_travelling_waves = [np.sum(PGD[i]>detection_threshold)/number_of_timesteps for i in range(len(PGD))]
```

```python
fig, ax = plt.subplots(1, figsize=(1.8,2.5), dpi=300)
c = plt.cm.cividis(0.95)
ax.boxplot(proportion_travelling_waves, positions=[0], notch=True, patch_artist=True,
           boxprops=dict(facecolor=c, color=c),
           capprops=dict(color=c),
           whiskerprops=dict(color=c),
           flierprops=dict(color=c, markeredgecolor=c),
           medianprops=dict(color=c)
          )
ax.set_xticklabels(['Gradient'], rotation=45)
ax.set_ylabel("Proportion of timepoints\n exhibiting travelling waves")
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.xlim([-0.2, 0.2])

plt.savefig(os.path.join(data_path, '01_results/checks/proportion-travelling-waves-boxplot_different-direction.png'), dpi=300, transparent=False, bbox_inches='tight')
plt.savefig(os.path.join(data_path, '01_results/checks/proportion-travelling-waves-boxplot_different-direction.svg'), dpi=300, transparent=False, bbox_inches='tight')
```

**Polar histogram of average phase directionality**

```python
# load data
n_trials = 100
phase_gradients = []

for i in range(n_trials):
    phase_gradients.append(np.load(os.path.join(data_path, f"checks/01_analysis_different-direction/phase_gradients_{i}.npy")))
    
direction = [-phase_gradients[i] for i in range(n_trials)]
```

```python
# compute the average direction across loactions and timepoints
avg_direction = np.array([np.mean(direction[i], axis=(0,1)) for i in range(n_trials)])

# compute polar angle for each trial
angles = np.angle(avg_direction[:,0]+1j*avg_direction[:,1])
```

```python
# polar histogram of phase-directions across trials
polar_histogram(angles, [], bin_width=2*np.pi/36, ylabel="Phasedirectionality", xticklabels=['x', 'y'], legend=['Gradient'])

plt.savefig(os.path.join(data_path, '01_results/checks/direction-polar-histogram_spatial-resolution.png'), dpi=300, transparent=False, bbox_inches='tight')
plt.savefig(os.path.join(data_path, '01_results/checks/direction-polar-histogram_spatial-resolution.svg'), dpi=300, transparent=False, bbox_inches='tight')
```

**Figure**. This figure shows a polar histogram of the average phase-directions (across timepoints and locations) across 100 trials for the gradient condition. This example shows again that the in-strength gradient determines the phase directionality.

### Results

**Polar histogram of average phase-direction**

```python
# load data
n_trials = 100
phase_gradients = []
phase_gradients_ctrl = []

for i in range(n_trials):
    phase_gradients.append(np.load(os.path.join(data_path, f"checks/01_analysis_high-frequency/phase_gradients_{i}.npy")))
    phase_gradients_ctrl.append(np.load(os.path.join(data_path, f"checks/02_analysis_high-frequency/phase_gradients_{i}.npy")))
    
direction = [-phase_gradients[i] for i in range(n_trials)]
direction_ctrl = [-phase_gradients_ctrl[i] for i in range(n_trials)]
```

```python
# compute the average direction across loactions and timepoints
avg_direction = np.array([np.mean(direction[i], axis=(0,1)) for i in range(n_trials)])
avg_direction_ctrl = np.array([np.mean(direction_ctrl[i], axis=(0,1)) for i in range(n_trials)])

# compute polar angle for each trial
angles = np.angle(avg_direction[:,0]+1j*avg_direction[:,1])
angles_ctrl = np.angle(avg_direction_ctrl[:,0]+1j*avg_direction_ctrl[:,1])
```

```python
# polar histogram of phase-directions across trials
polar_histogram(angles, angles_ctrl, bin_width=2*np.pi/36, ylabel="Phase directionality", xticklabels=['x', 'y'], legend=['Gradient', 'Uniform'])

plt.savefig(os.path.join(data_path, '01_results/checks/direction-polar-histogram_high-frequency.png'), dpi=300, transparent=False, bbox_inches='tight')
plt.savefig(os.path.join(data_path, '01_results/checks/direction-polar-histogram_high-frequency.svg'), dpi=300, transparent=False, bbox_inches='tight')
```

**Figure**. This figure shows a polar histogram of the average phase-directions (across timepoints and locations) across 100 trials for the gradient and control condition. While the control condition with uniform in-strength shows a uniform distribution of directions, the model with a hardcoded in-strength gradient shows a clear bias along the x-dimension that follows the in-strength gradient. 

**Polar histogram of average effective frequency gradient**

```python
# load data
n_trials = 100
frequency_gradients = []
frequency_gradients_ctrl = []

for i in range(n_trials):
    frequency_gradients.append(np.load(os.path.join(data_path, f"checks/01_analysis_high-frequency/frequency_gradients_{i}.npy")))
    frequency_gradients_ctrl.append(np.load(os.path.join(data_path, f"checks/02_analysis_high-frequency/frequency_gradients_{i}.npy")))
```
