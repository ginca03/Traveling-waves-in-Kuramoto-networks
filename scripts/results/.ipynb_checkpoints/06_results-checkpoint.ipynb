# Instrength vs. Effective frequency directed waves
Here, we test if effective frequency patterns mediate the effects of instrength gradients on wave direction or if instrength gradients directly control wave direction.

```python
import yaml
import pandas as pd
import numpy as np
import matplotlib.pylab as plt

from scipy import spatial
from scipy.stats import spearmanr, multivariate_normal

from tvb.simulator.lab import *

from mpl_toolkits.axes_grid1.anchored_artists import AnchoredSizeBar
import matplotlib.font_manager as fm

from palettable.cmocean.sequential import Ice_20
from palettable.scientific.sequential import Davos_20, LaPaz_20, Turku_20
import cmasher as cmr

cmap_potential = cmr.eclipse
cmap_waves = Ice_20.mpl_colormap
cmap_fq = LaPaz_20.mpl_colormap
cmap_strength = Davos_20.mpl_colormap
cmap_weights = Turku_20.mpl_colormap

c1 = '#3baa70'
c2 = '#703baa'

from modules.helpers import *
from modules.visualization import *
from modules.wave_detection_methods import *

dpi = 300
plt.rcParams.update({'font.size': 8})
```

```python
# Define functions
def exp_kernel(distance, sigma):
    """ Exponential kernel function for distance-dependent connection strength"""
    f = 1/(2*sigma) * np.exp(-distance/sigma)        
    return f
```

```python
# Read Configuration
try:
    configuration_path = "/Users/dk/Documents/Charite/PhD/travelingwaves/configuration/06_configuration.yaml"
    with open(configuration_path, "r") as ymlfile:
        config = yaml.load(ymlfile, Loader=yaml.BaseLoader)
except BaseException as e:
    print("Error: Specify correct path to yaml configuration file.")
    raise

    
# Simulation Parameters
# ---------------------
experiment_id = config["experiment_id"]
data_path = "/Users/dk/Documents/Charite/PhD/travelingwaves/data"

number_of_simulations = 100

integration_step_size = float(config["integration_step_size"])  # ms
simulation_duration = float(config["simulation_duration"])  # ms
initial_transient = float(config["initial_transient"])  # ms
initial_transient_samples = int(initial_transient/integration_step_size)  # samples
number_of_timesteps = int((simulation_duration-initial_transient)/integration_step_size)


# Model Parameters
# ----------------
intrinsic_frequency_mean = float(config["intrinsic_frequency_mean"])  # Hz - Kuramoto oscillator intrinsic frequency mean

coupling_strength = np.array(float(config["coupling_strength"]))  # global scaling of all connections
conduction_speed = float(config["conduction_speed"])  # mm/ms
connection_strength_scaling = float(config["connection_strength_scaling"])  # scaling factor for distance-dependent connection strength
connection_probability_scale = float(config["connection_probability_scale"])  # scale parameter for distance-dependent connection probability

instrength_gradient_offset = float(config["instrength_gradient_offset"])
instrength_gradient_scaling = float(config['instrength_gradient_scaling'])
sink_pos = np.array(config["sink_pos"])
source_pos = np.array(config["source_pos"])
singularity_width = float(config["singularity_width"])

# network parameters
nx = int(config["nx"])  # number of regions along x-dimension
ny = int(config["ny"])  # number of regions along y-dimension

N = nx*ny  # number of regions

x_extent = float(config["x_extent"])  # mm, extent of x-dimension
y_extent = float(config["y_extent"])  # mm, extent of y-dimension

# create node positions
y = np.linspace(0, y_extent, ny)
x = np.linspace(0, x_extent, nx)
pos = np.zeros((N,2))
pos[:,:] = np.array(np.meshgrid(x, y)).T.reshape(-1,2)

# prepare mesh
v = np.c_[pos, np.zeros(len(pos))]
f = spatial.Delaunay(pos).simplices
```

```python
### Aggregate flow potentials and effective frequency
U_sim = np.zeros((number_of_simulations, N))
effective_frequency_sim = np.zeros((number_of_simulations, N))

U_rho_sim = np.zeros(number_of_simulations)
U_rho_p_sim = np.zeros(number_of_simulations)
ef_rho_sim = np.zeros(number_of_simulations)
ef_rho_p_sim = np.zeros(number_of_simulations)

for sim_id in range(number_of_simulations):
    U = np.load(os.path.join(data_path, f"06_analysis/06_U_{sim_id}.npy"))
    U_sim[sim_id] = U.mean(axis=1)
    effective_frequency = np.load(os.path.join(data_path, f"06_analysis/06_effective_frequency_{sim_id}.npy"))
    effective_frequency_sim[sim_id] = np.mean(effective_frequency, axis=1)

    U_rho_sim[sim_id] = np.load(os.path.join(data_path, f"06_analysis/06_U_rho_{sim_id}.npy"))
    U_rho_p_sim[sim_id] = np.load(os.path.join(data_path, f"06_analysis/06_U_rho_p_{sim_id}.npy"))
    ef_rho_sim[sim_id] = np.load(os.path.join(data_path, f"06_analysis/06_ef_rho_{sim_id}.npy"))
    ef_rho_p_sim[sim_id] = np.load(os.path.join(data_path, f"06_analysis/06_ef_rho_p_{sim_id}.npy"))
    
U_avg = U_sim.mean(axis=0)
effective_frequency_avg = effective_frequency_sim.mean(axis=0)

U_rho_avg =  U_rho_sim.mean(axis=0)
U_rho_p_count = np.sum(U_rho_p_sim<=0.01)
ef_rho_avg = ef_rho_sim.mean(axis=0)
ef_rho_p_count = np.sum(ef_rho_p_sim<=0.01)

# Build dataframe
df = pd.DataFrame(np.column_stack([U_rho_sim, ef_rho_sim]), columns=['Instrength - flow potential correlation', 'Effective frequency - flow potential correlation'])
```

```python
# Plot correlations
sns.set_theme(style="ticks")

plt.figure(figsize=(2,2.))
plt.subplot(1,1,1)
p = sns.swarmplot(df, linewidth=0, palette=[c1, c2], size=4)
p.set_ylabel('Correlation')
p.set_yticks([-1, 0, 1])
p.set_xticklabels(['Instrength - \nflow potential', 'Effective frequency - \nflow potential'], rotation=35);

sns.despine()

#plt.savefig(os.path.join('/Users/dk/Documents/Charite/PhD/travelingwaves/data/06_results', f'06_correlations_swarmplot.png'), dpi=300, transparent=True, bbox_inches='tight')
#plt.savefig(os.path.join('/Users/dk/Documents/Charite/PhD/travelingwaves/data/06_results', f'06_correlations_swarmplot.svg'), dpi=300, transparent=True, bbox_inches='tight')
```

```python
# Print analyses stats
print(f'Average instrength - potential correlation: {U_rho_avg}')
print(f'Number of simulations with significant instrength - potential correlation: {U_rho_p_count}')
print(f'Average effective frequency - potential correlation: {ef_rho_avg}')
print(f'Number of simulations with significant effective frequency - potential correlation: {ef_rho_p_count}')
```
