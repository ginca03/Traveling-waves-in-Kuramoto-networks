# Subnetworks of full structural connectivity network

We ran nonnegative matrix factorizations and cross-validations in R (see scripts/analyses/nmf_sc.R)

```python
import pickle
import os
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
import pyvista as pv
from scipy.stats import spearmanr
from matplotlib.ticker import FormatStrFormatter
from scipy.cluster.hierarchy import dendrogram, linkage

from modules.helpers import correlation_spin_permutation_testing

from palettable.scientific.sequential import Turku_20, Davos_20

cmap_strength = Davos_20.mpl_colormap
cmap_weights = Turku_20.mpl_colormap
```

```python
data_path = "/Users/dk/Documents/Charite/PhD/travelingwaves/data"
figure_path = "/Users/dk/Documents/Charite/PhD/travelingwaves/data/48_results"
sphere_path = '/Users/dk/Documents/Charite/PhD/travelingwaves_old/data/connectomes/Schaefer2018_HCP_S900/positions_sphere.npy'
```

```python
# Load Mesh Geometry
# ------------------
# load schaefer surface mesh
with open(os.path.join(data_path, "connectomes/Schaefer2018_HCP_S900/schaefer_surface_mesh.pkl"), 'rb') as f:
    surface_mesh = pickle.load(f)
    
v_lh = surface_mesh['vertices_lh']
f_lh = surface_mesh['faces_lh']
v_rh = surface_mesh['vertices_rh']
f_rh = surface_mesh['faces_rh']
number_of_regions_per_hemi = v_lh.shape[0]
number_of_regions = number_of_regions_per_hemi * 2

# combine left and right hemispheres
v_lh[:,0] = v_lh[:,0] - (v_lh[:,0].max() - v_lh[:,0].min())

v = np.concatenate([v_lh, v_rh])
v[:,0] = v[:,0]
f = np.concatenate([f_lh, f_rh+number_of_regions_per_hemi])
```

### Create subject SC data matrix

```python
# load data
weights_nsc = np.load(os.path.join(data_path, 'connectomes/Schaefer2018_HCP_S900/weights_nsc.npy'))
avg_weights = np.load(os.path.join(data_path, 'connectomes/Schaefer2018_HCP_S900/avg_weights.npy'))
```

```python
# Upper triangular matrix
# create consistency-thresholded mask
weights_mask = avg_weights!=0
weights_mask[np.triu_indices_from(avg_weights, k=1)] = 0

# put upper triangular subjects SCs in columns of datamatrix.
#weights_triu = np.array([weights_nsc[i][weights_mask] for i in range(len(weights_nsc))])
#
#np.save("weights_triu_consensus.npy", weights_triu)
weights_triu = np.load(os.path.join(data_path, "48_results/weights_triu_consensus.npy"))

number_of_subjects = weights_triu.shape[0]

print(weights_triu.shape)
```

### Nonnegative Matrix Factorization Cross-Validation

```python
ranks = np.arange(2,11)
number_of_runs = 100

# Get MSE of cross-validation from R and convert to RMSE
cv_mse = np.load(os.path.join(data_path, "48_results/cv_impute_mse.npy"))
cv_rmse = pd.DataFrame(np.sqrt(cv_mse.reshape((number_of_runs, len(ranks)))), columns=ranks)  # create CV RMSE dataframe
```

```python
# Plot CV RMSE
fig, ax = plt.subplots(figsize=(3.5, 3))

sns.pointplot(cv_rmse, ax=ax, color='#29528a', errorbar='sd')
plt.xlabel('Number of subnetworks')
plt.ylabel('Root mean squared error (a.u.)')
sns.despine()

ax2 = plt.axes([0.3, 0.6, .35, .3])
sns.pointplot(cv_rmse.iloc[:,:5], ax=ax2, color='#29528a', errorbar='sd')
ax2.set_xlim([-0.5, 3.5])
ax2.set_ylim([99., 102.])
sns.despine()

#plt.savefig(os.path.join(figure_path, f'48_cv_rmse.svg'), dpi=300, transparent=False, bbox_inches='tight')
#plt.savefig(os.path.join(figure_path, f'48_cv_rmse.png'), dpi=300, transparent=False, bbox_inches='tight')
```

```python
# Identify best number of subnetworks using RMSE
best_rank = cv_rmse.mean(axis=0).idxmin()
print('Best rank: %i' % best_rank)
```

### Reconstruct subnetworks from best NMF

```python
# Get weights, components and scaler from R
W = np.load(os.path.join(data_path, "48_results/nmf_components.npy"))
H = np.load(os.path.join(data_path, "48_results/nmf_weights.npy"))
d = np.load(os.path.join(data_path, "48_results/nmf_scaler.npy"))
```

```python
# reconstruct full data matrix
A = W.dot(np.tile(d, (number_of_subjects,1)).T*H)

print('RMSE: %.2f' % np.sqrt(np.mean((weights_triu - A.T)**2)))
print('Explained variance: %.2f %%' % (100*(1 - np.sum((weights_triu - A.T)**2) / np.sum((weights_triu - np.mean(weights_triu))**2))))
```

```python
# reconstruct subnetworks
n_components = best_rank
components = []

for i in range(n_components):
    A_comp = np.outer(W[:,i], d[i]*H[i])

    recon = np.zeros_like(avg_weights)
    recon[weights_mask] = np.mean(A_comp, axis=1)
    components.append(recon + recon.T)
```

```python
### Plot subnets instrengths ###
pv.set_plot_theme('document')

# Build pyvista mesh
mesh = pv.PolyData(v)

# convert faces to pyvista
mesh.faces = np.insert(f.flatten(), np.arange(0,len(f.flatten()),3), 3)

p = pv.Plotter(shape=(3,2), off_screen=True, notebook=True, border=False, window_size=[200,350], lighting='none')
p.enable_anti_aliasing('msaa')

for i, sn in enumerate(range(n_components)):
    p.subplot(*np.unravel_index(i, (3,2)))
    mesh.point_data['data'] = components[i].sum(axis=0)
    
    p.add_mesh(mesh, point_size=0, show_scalar_bar=False, smooth_shading=True, cmap=cmap_strength, copy_mesh=True)
    p.add_text(str(sn), font='arial', font_size=14, position=(35,210))
    
    p.camera_position = 'xy'
    p.camera.azimuth = 0
    p.camera.elevation = 0
    p.camera.roll = 0
    p.camera.zoom(1.5)

p.show()
#p.screenshot(os.path.join(figure_path, "rank5_subnetworks_instrength.png"), transparent_background=True)
p.close()
```

```python
# Identify components that are most strongly correlated with the anterior-posterior position
antpos_rho = []
for i in range(len(components)):
    antpos_rho.append(spearmanr(v[:,1], components[i].sum(axis=0))[0])

alpha_idx = np.argmax(antpos_rho)
beta_idx = np.argmin(antpos_rho)

alpha_sc = components[alpha_idx]
beta_sc = components[beta_idx]

# save frequency-specific SCs
#np.save("/Users/dk/Documents/Charite/PhD/travelingwaves/data/hcp_meg/nmf_weights_alpha.npy", alpha_sc)
#np.save("/Users/dk/Documents/Charite/PhD/travelingwaves/data/hcp_meg/nmf_weights_beta.npy", beta_sc)

print("Alpha component: %i" % np.argmax(antpos_rho))
print("Beta component: %i" % np.argmin(antpos_rho))
```

```python
# Explained variance of combined alpha and beta subnetworks
A_recon = W[:,[alpha_idx, beta_idx]] @ np.diag(d[[alpha_idx, beta_idx]]) @ H[[alpha_idx, beta_idx]]
expvar_recon = 1 - np.sum((weights_triu - A_recon.T)**2) / np.sum((weights_triu - np.mean(weights_triu))**2)

print('Explained variance: %.2f %%' % (100*expvar_recon))
```

```python
# Original instrength vs. subnetwork instrength correlation
instrength = avg_weights.sum(axis=0)

correlation_spin_permutation_testing((alpha_sc + beta_sc).sum(axis=0), instrength, sphere_path=sphere_path)
```

### Across-rank similarity

```python
# load nmf components across ranks from R
W_all = np.load(os.path.join(data_path, "48_results/nmf_multirank_factors.npy")).T

# Compute across rank correlation
rhom = np.corrcoef(W_all)
```

```python
# visualize dendrogram based on component correlation
dendro = dendrogram(linkage(rhom, method='ward'), no_plot=True)

cm = sns.clustermap(rhom, method='ward', dendrogram_ratio=0.22, tree_kws=dict(linewidths=1.5), cbar_kws={"ticks":[rhom.min(),1], "label": "Correlation"}, figsize=(5,5))
cm.ax_heatmap.set_yticklabels(cm.ax_heatmap.get_yticklabels(), size = 12, rotation=0)
cm.ax_heatmap.set_xticklabels(cm.ax_heatmap.get_xticklabels(), size = 12, rotation=90)
cm.ax_cbar.tick_params(labelsize=12)
cm.ax_cbar.yaxis.label.set_size(14)
cm.ax_cbar.yaxis.set_major_formatter(FormatStrFormatter('%.2f'))

#cm.figure.savefig('/Users/dk/Documents/Charite/PhD/travelingwaves/data/48_results/clustermap.png', dpi=300)
#cm.figure.savefig('/Users/dk/Documents/Charite/PhD/travelingwaves/data/48_results/clustermap.svg', dpi=300)
```

```python
# Reconstruct subnetworks
components_all = []
for i in range(len(W_all)):
    recon = np.zeros_like(avg_weights)
    recon[weights_mask] = W_all[i]
    components_all.append(recon + recon.T)

subnet = [np.array(dendro['leaves'])[np.array(dendro['leaves_color_list'])=='C1'],
          np.array(dendro['leaves'])[np.array(dendro['leaves_color_list'])=='C2']]
```
